Exemple 3.3

```python
import pandas as pd 
import numpy as np

s = pd.DataFrame([1, 2, 3, np.NaN, 5, 6, None])

df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', None],
    'style': ['cup', None, 'cup', 'pack', 'pack'],
    'rating': [4, 4, np.NaN, 15, 5]})

print("Visualiser les données manquantes : true means a detection of a missing data")
print(df.isnull())

print("\nIsoler les données manquantes")
print(df[df.isnull().any(axis=1)])

print("\nNombre des données manquantes par colonne:")
print(df.isnull().sum())
print("Nombre total de données manquantes:")
print(df.isnull().sum().sum())

print("\nRemplacer les données manquantes par la moyenne")
df_filled = df.copy()
df_filled['rating'] = df['rating'].fillna(df['rating'].mean())
print(df_filled)

print("\nSupprimer les données manquantes")
df_dropped = df.dropna()
print(df_dropped)
```

Exemple 3.4

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Create an empty dataset
df = pd.DataFrame()

# Create two variables called x0 and x1. Make the first value of x1 a missing value
df['x0'] = [0.3051,0.4949,0.6974,0.3769,0.2231,0.341,0.4436,0.5897,0.6308,0.5]
df['x1'] = [np.nan,0.2654,0.2615,0.5846,0.4615,0.8308,0.4962,0.3269,0.5346,0.6731]

# View the dataset
print("Data :")
print(df)

# Create an imputer object that looks for 'Nan' values, then replaces them with the mean value of the feature by columns (axis=0)
print("\nChercher les valeurs manquantes par la moyenne de la colonne")
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Apply the imputer on the df dataset
imputed_df = pd.DataFrame(mean_imputer.fit_transform(df), columns=df.columns)

# View the dataset
print("Data après imputation par la moyenne :")
print(imputed_df)

# Change the imputation strategy to median
print("\nChercher les valeurs manquantes par la médiane de la colonne")
median_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputed_df_median = pd.DataFrame(median_imputer.fit_transform(df), columns=df.columns)
print("Data après imputation par la médiane :")
print(imputed_df_median)

# Change the imputation strategy to most_frequent
print("\nChercher les valeurs manquantes par la valeur la plus fréquente de la colonne")
most_frequent_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputed_df_most_frequent = pd.DataFrame(most_frequent_imputer.fit_transform(df), columns=df.columns)
print("Data après imputation par la valeur la plus fréquente :")
print(imputed_df_most_frequent)
```

Exemple 3.5

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'A': [2,1,2,3,3,5,4],'B': [1,2,3,5,4,2,5],
                   'C': [5,3,4,1,1,2,3]})
print("Data originale :\n")
print(df)

# Trier le DataFrame selon la colonne 'A'
print("\nLe jeu de données est trié selon la colonne 'A':")
df_sorted = df.sort_values(by='A').reset_index(drop=True)
print(df_sorted)

# Mélanger le DataFrame
print("\nLe jeu de données est mélangé:")
index = df.index.tolist()
np.random.shuffle(index)
df_shuffled = df.iloc[index].reset_index(drop=True)
print(df_shuffled)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
from sklearn.preprocessing import MinMaxScaler
 
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/cpu-performance/machine.data'
 
names = ['constructor','Model','MYCT','MMIN','MMAX','CACH','CHMIN','CHMAX','PRP','ERP']
 
dataset = pd.read_csv(url, names=names)

print("data:")
print(dataset.head())  # afficher uniquement les premières lignes pour une meilleure lisibilité
 
# Appliquer MIN MAX SCALING sur MYCT et MMAX
data_to_scale = dataset[['MYCT', 'MMAX']]
minmax_scale = MinMaxScaler().fit(data_to_scale)
df_minmax = minmax_scale.transform(data_to_scale)
 
print('\n********** Normalisation*********\n')
 
print('Moyenne après le Min max Scaling :\nMYCT={:.2f}, MMAX={:.2f}'
.format(df_minmax[:,0].mean(), df_minmax[:,1].mean()))
 
print('\n')
 
print('Valeur minimale et maximale pour la feature MYCT après min max scaling: \nMIN={:.2f}, MAX={:.2f}'
.format(df_minmax[:,0].min(), df_minmax[:,0].max()))
 
print('\n')
 
print('Valeur minimale et maximale pour la feature MMAX après min max scaling : \nMIN={:.2f}, MAX={:.2f}'
.format(df_minmax[:,1].min(), df_minmax[:,1].max()))

print("\ndata après normalisation")
print(pd.DataFrame(df_minmax, columns=['MYCT_normalized', 'MMAX_normalized']).head())
```

Exemple 3.7

```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import StandardScaler  # Correct the import

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/cpu-performance/machine.data'
 
names= ['constructor','Model','MYCT','MMIN','MMAX','CACH','CHMIN','CHMAX','PRP','ERP']
 
dataset = pd.read_csv(url, names=names)
print("data:")
print(dataset.head())  # Just displaying the head for brevity
 
# Z-Score standardisation 
std_scaler = StandardScaler().fit(dataset[['MYCT', 'MMAX']])
df_std = std_scaler.transform(dataset[['MYCT', 'MMAX']])
 
print('\n********** Standardisation*********\n')

print('Moyenne et Ecart type apres la standardisation de la feature MYCT :\nMoyenne = {:.2f}, Ecart-Type = {:.2f}'
      .format(df_std[:,0].mean(), df_std[:,0].std()))

print('\nMoyenne et Ecart type apres la standardisation de la feature MMAX :\nMoyenne = {:.2f}, Ecart-Type = {:.2f}'
      .format(df_std[:,1].mean(), df_std[:,1].std()))
 
print('\nValeur minimale et maximale pour la feature MYCT apres standardisation: MIN = {:.2f}, MAX = {:.2f}'
      .format(df_std[:,0].min(), df_std[:,0].max()))

print('\nValeur minimale et maximale pour la feature MMAX apres standardisation: MIN = {:.2f}, MAX = {:.2f}'
      .format(df_std[:,1].min(), df_std[:,1].max()))

print('\nAffichage des valeurs apres scaling')
print(pd.DataFrame(df_std, columns=['MYCT_standardized', 'MMAX_standardized']).head())
```
